```python
from keras.models import load_model
import numpy as np
import librosa
import os
import json
        
```

```python
# Load the saved model
model = load_model('/home/lab/SP-testing/Machine Learning Model/CNN/CNN.h5')

with open('model_and_classes.json', 'r') as json_file:
    class_label_mapping = json.load(json_file)

# Specify the path to the WAV file for prediction
directory_path = '/home/lab/SP-testing/Machine Learning Model/CNN/Outtest/'



```

```python
for filename in os.listdir(directory_path):
    if filename.endswith('.wav'):  # Make sure the file is a WAV file
        file_path = os.path.join(directory_path, filename)
        
        # Load and preprocess the audio data
        y, sr = librosa.load(file_path, sr=None)
        y = y.reshape(1, -1)
        
        # Perform the prediction
        predictions = model.predict(y)  # Assuming the model expects batches of data
        print(np.max(predictions))
        print(predictions)
        # Get the predicted class index
        predicted_class_index = np.argmax(predictions)
        
        # Map the predicted numerical label to the corresponding class name using the class label mapping
        predicted_class_name = None
        for class_name, label in class_label_mapping.items():
            if label == predicted_class_index:
                predicted_class_name = class_name
                break
        
        print(f"File: {filename}, Predicted class: {predicted_class_name}")
```
